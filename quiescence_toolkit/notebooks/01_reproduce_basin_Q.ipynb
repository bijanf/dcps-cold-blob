# Notebook 01 — Reproduce the basin-scale Quiescence Index Q

Goal: reproduce the published Q values for the North Atlantic,
North Pacific, and Antarctic Circumpolar Current sectors
(Fallah & Rostami, in preparation).

Inputs (all in the parent `dcps-cold-blob` repository):
  - ORAS5 SST + SSH (loaded via `dcps.products.load_oras5_var`)
  - Pre-computed cache `dcps/cache/multi_basin/multi_basin_quiescence.json`

Expected outputs:
  - Q_NA ≈ 0.32 ± 0.03
  - Q_NP ≈ 0.35 ± 0.04
  - Q_ACC ≈ 0.46 ± 0.05

Acceptance criterion: each Q within ±0.05 of the published value.

```python
import json
import sys
from pathlib import Path

import numpy as np

# Path to the parent dcps-cold-blob repo
REPO = Path('/home/bijanf/Documents/NEW_Theory')
sys.path.insert(0, str(REPO / 'dcps'))
sys.path.insert(0, str(REPO / 'quiescence_toolkit' / 'src'))

from quiescence_index import Q_from_basin_cache

PUBLISHED = {'atlantic': 0.32, 'pacific': 0.35, 'southern': 0.46}
TOL = 0.05

for basin in ('atlantic', 'pacific', 'southern'):
    result = Q_from_basin_cache(basin)
    if result is None:
        print(f'{basin}: cache not available')
        continue
    q = result['Q']
    delta = q - PUBLISHED[basin]
    pass_fail = 'PASS' if abs(delta) <= TOL else 'FAIL'
    print(f'{basin:<10} Q = {q:+.3f} (published {PUBLISHED[basin]:+.3f}, '
          f'delta = {delta:+.3f}) -- {pass_fail}')
```
